# Initiating Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
X = sales.drop(['Branch', 'City', 'Product line', 'Rating'], axis=1)
y = sales['Rating'] 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def cross_val(model):
    pred = cross_val_score(model, X, y, cv=10)
    return pred.mean()

def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2_Square', r2_square)
    
    
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('stdscalar', StandardScaler())
])

X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression(normalize=True)
lin_reg.fit(X_train,y_train)
```

```python
print(lin_reg.intercept_)
```

```python
coeff_sales = pd.DataFrame(lin_reg.coef_, X.columns, columns=['Coefficient'])
coeff_sales
```

```python
pred = lin_reg.predict(X_test)
```

```python
pred
```

```python
test_pred = lin_reg.predict(X_test)
train_pred = lin_reg.predict(X_train)

print('Test data evaluation:\n')
print_evaluate(y_test, test_pred)
print('Train data evaluation:\n')
print_evaluate(y_train, train_pred)

results_df = pd.DataFrame(data=[["Linear Regression", *evaluate(y_test, test_pred) , cross_val(LinearRegression())]], 
                          columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Square', "Cross Validation"])
```
